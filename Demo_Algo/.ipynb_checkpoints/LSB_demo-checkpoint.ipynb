```python
np_img = np.array(image)
img_width, img_height, rgb = np_img.shape
np_img, img_width, img_height
```

```python
def txt_to_bits(path):
    with open(path, 'r', encoding='utf-8') as file:
        message = file.read()
    message = message + "\0\0"
    message_bytes = message.encode('utf-8')
    message_bits = ''.join(format(byte, '08b') for byte in message_bytes)
    
    return message_bits

def bits_to_txt(message_bits, path='decoded.txt'):
    
    message_bytes = bytearray(int(message_bits[i:i+8], 2) for i in range(0, len(message_bits), 8)) 
    message = message_bytes.decode('utf-8')

    with open(path, 'w', encoding='utf-8') as file:
        file.write(message)

path = 'test_message.txt'
bits_to_txt(txt_to_bits(path))

with open('decoded.txt', 'r', encoding='utf-8') as file:
    file_mess = file.read()

print(file_mess)
```

```python

message_bits = txt_to_bits(path)
np_img = np_img.flatten()

for idx, bit in enumerate(message_bits):
    val = bin(np_img[idx])[2:].zfill(8)
    val = val[:-1] + bit
    np_img[idx] = int(val,2)

np_img = np_img.reshape(img_width,img_height,3)
np_img, plt.imshow(np_img)
```

```python


# Flatten the image array
np_img = np_img.flatten()

# Initialize variables
secret_bits = ''
idx = 0

# Define your end signature if used
signature = '00000000'  # Example for 8-bit end marker, adjust if needed

# Extract bits from the last bit of each channel
while len(secret_bits) < 8 or secret_bits[-8:] != signature:  # Adjust based on signature length
    # Get the last bit from each of the RGB channels
    for _ in range(3):  # 3 channels (R, G, B)
        if idx >= len(np_img):
            break  # Prevent index out of bounds
        val = bin(np_img[idx])[2:].zfill(8)  # Convert to 8-bit binary
        secret_bits += val[-1]  # Append the last bit
        idx += 1
    if idx >= len(np_img):
        break  # Prevent index out of bounds

# Remove the signature (last 8 bits)
if len(secret_bits) >= 8:
    secret_bits = secret_bits[:-8]  # Remove the end marker bits

# Print out the bits of the secret message
print("Secret bits:", secret_bits)

# Optionally convert bits to characters if needed
decoded_msg = ''.join(chr(int(secret_bits[i:i+8], 2)) for i in range(0, len(secret_bits), 8))
print("Decoded message:", decoded_msg)
```
